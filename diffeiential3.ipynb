```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
```

```python
# 微分方程式を定義
def odeSIR(t, y, beta, gamma):
    dS = -beta*y[0]*y[1]
    dI = beta*y[0]*y[1] -gamma*y[1]
    dR = gamma*y[1]
    return[dS, dI, dR]
# パラメータ
beta, gamma = 0.01, 0.5
Tend = 10.0
coef = 3
S0 = coef*gamma/beta
I0 = 10
R0 = 0

# 微分方程式を解く
sol = solve_ivp(fun=odeSIR, t_span=[0, Tend], y0=[S0, I0, R0], args=[beta, gamma], dense_output=True)
```

```python
tc = np.linspace(0, Tend, 100)
yc = sol.sol(tc)

plt.plot(tc, yc[0], label="S(t): susceptibles" )
plt.plot(tc, yc[1], label="I(t): infectives")
plt.plot(tc, yc[2], label="R(t): recovered")
plt.xlabel("time")
plt.ylabel("number")
plt.legend()
plt.grid()
```
