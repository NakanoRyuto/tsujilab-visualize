```python
# アニメーションを作る
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from IPython.display import HTML 
```

```python
fig = plt.figure()
img_buffer = []

x = 0 #初期状態のX座標
y = 0 #初期状態のX座標

plt.xlim([0, 10])
plt.ylim([-5, 5])

for t in range(30):
    x = x + 1
    if x > 8: # x==8の壁でストップする
        x = x - 1
    y = y
    img = plt.plot(x, y, linewidth=0, marker='o', markersize=15, color='r')
    img_buffer.append(img)

ts = 100 #アニメーションの間隔
ani = animation.ArtistAnimation(fig, img_buffer, interval=ts, repeat=False)

plt.close()
HTML(ani.to_jshtml())
```

```python
#株価のシュミレーション

x = []
y = []

xt = 1000 #株価の初期値

for t in range(100):
    xt = xt + np.random.rand() * 200 - 100 #ランダムで -100 ~ 100 加算する
    x.append(t)
    y.append(xt)
    
plt.plot(x, y, marker='o')
```
