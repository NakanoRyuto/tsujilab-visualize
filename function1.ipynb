```python
#　関数のグラフを作ってみよう

import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 4, 5)
y = x 

plt.plot(x, y, marker='o')
```

```python

x = np.linspace(0, 4, 5)
y = x + 2
 
plt.plot(x, y, marker='o')
```

```python
x = np.linspace(0, 4, 5)
y = x + 2

plt.xlim([0, 6])
plt.ylim([0, 6])
 
plt.plot(x, y, marker='o')
```

```python
x = np.linspace(0, 4, 5)
y = 2 * x

plt.xlim([0, 6])
plt.ylim([0, 6])
 
plt.plot(x, y, marker='o')
```

```python
x = np.linspace(0, 4, 5)
y = 3 * x

plt.xlim([0, 6])
plt.ylim([0, 6])
 
plt.plot(x, y, marker='o')
```

```python
x = np.linspace(0, 4, 5)
y = -x

plt.xlim([0, 6])
plt.ylim([0, -6])
 
plt.plot(x, y, marker='o')
```

```python
x = np.linspace(0, 4, 5)
y = -x + 3

plt.xlim([0, 6])
plt.ylim([0, 6])
 
plt.plot(x, y, marker='o')
```

```python
x = np.linspace(0, 4, 5)
y = x ** 3

plt.xlim([0, 6])
plt.ylim([0, 18])
 
plt.plot(x, y, marker='o')
```
